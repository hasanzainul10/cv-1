```python
import cv2 as cv
import matplotlib.pyplot as plt
# splitting and merging color channel
img = cv.imread(r'C:\Users\user\Documents\GitHub\cv-1\essential\assets\ca.jpg')
nrows = 2
ncols = 2
font_size=10
plt.figure(figsize=(15,10))

# img with all channel intact
imgRGB = img[:,:,::-1]

# show_in_matplotlib(imgRGB)
plt.subplot(nrows,ncols,1)
plt.title('Original',fontsize=font_size)
plt.imshow(imgRGB)
plt.axis('off')



# Image without blue channel
imgNoBlue = img.copy()
imgNoBlue[:,:,0]=0
plt.subplot(nrows,ncols,2)
plt.title('without blue channel',fontsize=font_size)
imgRGB = imgNoBlue[:,:,::-1] 
plt.imshow(imgRGB)
plt.axis('off')



# Image without green channel
imgNoGreen = img.copy()
imgNoGreen[:,:,1]=0
plt.subplot(nrows,ncols,3)
plt.title('without green channel',fontsize=font_size)
imgRGB = imgNoGreen[:,:,::-1] 
plt.imshow(imgRGB)
plt.axis('off')



# Image without red channel
imgNoRed = img.copy()
imgNoRed[:,:,2]=0
plt.subplot(nrows,ncols,4)
plt.title('without red channel',fontsize=font_size)
imgRGB = imgNoRed[:,:,::-1] 
plt.imshow(imgRGB)
plt.axis('off')


plt.show()
```

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

path = r'../essential/assets/ca.jpg'
img = cv.imread(path)
img = cv.cvtColor(img,cv.COLOR_RGB2BGR)


kernel_edge_1 = np.array([[1, 0, -1],
                        [0, 0, 0], 
                        [-1, 0, 1]])

kernel_edge_2 = np.array([[0, 1, 0],
                        [1, -4, 1],
                        [0, 1, 0]])

kernel_edge_3 = np.array([[-1, -1, -1],
                        [-1, 8, -1],
                        [-1, -1, -1]])

sobel_x_kernel = np.array([[1, 0, -1],
                           [2, 0, -2],
                           [1, 0, -1]])

sobel_y_kernel = np.array([[1, 2, 1],
                           [0, 0, 0],
                           [-1, -2, -1]])

img_kernel_edge_1 = cv.filter2D(img, -1, kernel_edge_1)
img_kernel_edge_2 = cv.filter2D(img, -1, kernel_edge_2)
img_kernel_edge_3 = cv.filter2D(img, -1, kernel_edge_3)
img_sobel_x_kernel = cv.filter2D(img, -1, sobel_x_kernel)
img_sobel_y_kernel = cv.filter2D(img, -1, sobel_y_kernel)



img_list = [img, img_kernel_edge_1, img_kernel_edge_2, img_kernel_edge_3, img_sobel_x_kernel, img_sobel_y_kernel]


nrows = 2
ncols = 3
font_size = 10
plt.figure(figsize=(30,15))


for i in range(len(img_list)):
    
    plt.subplot(nrows,ncols,i+1)
    
    if i == 0:
        plt.title('Original',fontsize=font_size)
    elif i == 1:
        plt.title('kernel_edge_1',fontsize=font_size)
    elif i == 2:
        plt.title('kernel_edge_2',fontsize=font_size)
    elif i == 3:
        plt.title('kernel_edge_3',fontsize=font_size)
    elif i == 4:
        plt.title('sobel_x_kernel',fontsize=font_size)
    elif i == 5:
        plt.title('sobel_y_kernel',fontsize=font_size)

    plt.imshow(img_list[i])
    plt.axis('off')

plt.show()
```
