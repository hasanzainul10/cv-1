```python
# Color Dictionary

# define color constants
# BGR format
blue = (255,0,0)
green = (0,255,0)
red = (0,0,255)
yellow = (0,255,255)
magenta = (255,0,255)
dark_gray = (50,50,50)
light_gray = (200,200,200)
dark_blue = (200,100,0)


color = {
    'Blue':blue,
    'Green':green,
    'Red':red,
    'Yellow':yellow,
    'Magenta':magenta,
    'Dark_gray':dark_gray,
    'Dark_blue':dark_blue
    

}
```

```python
# Functions
import numpy as np
import matplotlib.pyplot as plt

def create_canvas(height = 500, width= 500, bgColor = 0):
    canvas = np.zeros((height,width,3),dtype='uint8')
    if bgColor!=0:
        canvas[:] = bgColor
    return canvas


def show_in_matplotlib(img, figsize=(6,8)):
    imgRGB = cv.cvtColor(img, cv.COLOR_BGR2RGB)
    plt.figure(figsize=figsize)
    
    plt.imshow(imgRGB)
    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv


img = create_canvas(height=500,width=500,bgColor=color['Dark_gray'])


triangle = np.array([[250,175],[185,288],[315,288]])

cv.polylines(img, [triangle], True, color['Blue'], 3)

show_in_matplotlib(img)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv


img = create_canvas(height=500,width=500,bgColor=color['Dark_gray'])


hexagon = np.array([[288,185],[213,185],[175,250],[213,315],[287,315],[325,250]])

cv.polylines(img, [hexagon], True, color['Yellow'], 3)

show_in_matplotlib(img)
```
