Extract a smaller dimension of 3d data

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(9)
# create 3d
a3d = np.random.randint(0,256,(100,100,3), dtype='uint8')

# show original
plt.imshow(a3d)
plt.show()
```

```python
# extract 10x10x3
a3d_10x10 = a3d[0:11,0:11,:]

# show cropped image
plt.imshow(a3d_10x10)
plt.show()
```

```python
# extract centre of the array
a3d_centre = a3d[40:61,40:61,:]
plt.imshow(a3d_centre)
plt.show()
```

```python
a00 = a3d_centre[-1,-1,:].copy()
a00 = a00.reshape(1,1,3)
plt.imshow(a00)
plt.show()
```

```python
##
A = np.array([
    [2,4],
    [1,-2],
    ])

b = np.array([
    [4],
    [2],
    ])
invA = np.linalg.inv(A)

x=invA @ b
print(x)
```

```python
# check if correct from equation
2*x[0]+4*x[1]==4
```

```python
# alternative way

np.linalg.solve(A,b)
```

```python
# put left side of eq(put all variable here)
A = np.array([
    [2,4,1],
    [1,-24,5.5],
    [2,-4.2,1.5]
])

# put answer here
b= np.array([
    [4],
    [2],
    [-2]
])

# method one
invA = np.linalg.inv(A)

# @ is matrix multiplication
x=invA @ b
print(x)
# method 2
np.linalg.solve(A,b)
```

#### Reversing or sorting array using indexing

```python
# Reversing or sorting array using indexing

import numpy as np

a = np.arange(10)

print(a)
```

```python
## accesss last value
a[-1]
```

```python
## get last 2 value
a[-2:]
```

```python
# get odd value
# start at 1 and step 2 value
a[1::2]
```

```python
# positive step
a[0::1]

```

```python
# negative step
# moves iteration backward
a[::-1]
```

```python


# 
a[0::-1]
```
