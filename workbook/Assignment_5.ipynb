```python
##INHERITANCE##
class Student:
    def __init__(self,name,matricNo,course):
        self.name =name
        self.matricNo=matricNo
        self.course = course
    
    def printInfo(self):
        print(f"Student name : {self.name}")
        print(f"{self.name} matric no : {self.matricNo}")
        print(f"{self.name} courses : {self.course}")



class BSCStudent(Student):
    def __init__(self, name, matricNo, major,course,passed,failed):
        super().__init__(name, matricNo, course)
        self.avg=0
        self.major = major
        self.course = course
        self.passed = passed
        self.failed = failed
    
    def average(self):
        l1 =list(self.passed.values())
        l2 = list(self.failed.values())
        l3 = l1+l2
        l3 = list(l3)
        print(l3)
        sum = 0
        for i in range(len(l3)):
            sum+=l3[i]
        print(sum)
        self.avg = sum/len(l3)
        # print(avg)
        return self.avg
    
    def requiredCourses(self):
        self.retake = self.course.intersection(self.failed)
        return self.retake
    
    
    def printInfo(self):
        avg = self.average()
        retake = self.requiredCourses()
        super().printInfo()
        print(f"{self.name} Major : {self.major}")
        print(f"{self.name}'s Major courses : {self.course}")
        print(f"{self.name}'s Passed courses : {self.passed}")
        print(f"{self.name}'s failed courses : {self.failed}")
        print(f"Subject need to be taken by {self.name} : {retake}")
        print(f"{self.name}'s Average score : {avg}")



            




bs = BSCStudent("Hasan",17144133,"is",{"algo","is","database","os"}, {"math":80,"english":90},{"ai":50,"algo":40})

bs.printInfo()
```
