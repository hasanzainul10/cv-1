```python
##INHERITANCE##
class student:
    def __init__(self,name,matricNo,course):
        self.name =name
        self.matricNo=matricNo
        self.course = course
    
    def printInfo(self):
        print(f"Student name : {self.name}")
        print(f"Student matric no : {self.matricNo}")
        print(f"Student course : {self.course}")



class BSCstudent(student):
    def __init__(self, name, matricNo, major,course,passed,failed):
        super().__init__(name, matricNo, course)
        self.avg=0
        self.major = major
        self.course = course
        self.passed = passed
        self.failed = failed
    
    def average(self):
        l1 =list(self.passed.values())
        l2 = list(self.failed.values())
        l3 = l1+l2
        print(l3)
        sum = 0
        for i in range(len(l3)):
            sum+=l3[i]
        print(sum)
        avg = sum/len(l3)
        # print(avg)
        return avg
    
    def requiredCourses(self):
        return self.course
    
    
    def printInfo(self):
        avg = self.average()
        super().printInfo()
        print(f"Major : {self.major}")
        print(f"Major courses : {self.course}")
        
        print(f"Average score : {avg}")



            




bs = BSCstudent("Hasan",17144133,"is",["algo","is","database","os"], {"math":80,"english":90},{"ai":50,"algo":40})

bs.printInfo()
```
