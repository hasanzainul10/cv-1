```python
# 1. Apply thresholding
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
plt.figure(figsize=(12,10))
plt.axis('off')

path = r"essential\assets\ca.jpg"
img = cv.imread(path) 

imgGray = cv.cvtColor(img,cv.COLOR_BGR2GRAY)
thresh = 100
maxValue = 255

ret, BinaryThresh = cv.threshold(imgGray, thresh, maxValue, cv.THRESH_BINARY_INV)
ret, TruncateThresh = cv.threshold(imgGray, thresh, maxValue, cv.THRESH_TRUNC)
ret, ToZeroThresh = cv.threshold(imgGray, thresh, maxValue, cv.THRESH_TOZERO)
ret, ToZeroInvThresh = cv.threshold(imgGray, thresh, maxValue, cv.THRESH_TOZERO_INV)



plt.subplot(221)
plt.imshow(cv.cvtColor(BinaryThresh,cv.COLOR_BGR2RGB))
plt.title(f'Binary thresh={thresh}, maxval={maxValue}')
plt.axis('off')

plt.subplot(222)
plt.imshow(cv.cvtColor(TruncateThresh,cv.COLOR_BGR2RGB))
plt.title(f'Truncate thresh={thresh}, maxval={maxValue}')
plt.axis('off')

plt.subplot(223)
plt.imshow(cv.cvtColor(ToZeroThresh,cv.COLOR_BGR2RGB))
plt.title(f'Tozero thresh={thresh}, maxval={maxValue}')
plt.axis('off')

plt.subplot(224)
plt.imshow(cv.cvtColor(ToZeroInvThresh,cv.COLOR_BGR2RGB))
plt.title(f'Tozero Inv thresh={thresh}, maxval={maxValue}')
plt.axis('off')


plt.show()

```

```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
plt.figure(figsize=(10,8))

path = r"essential\assets\ca.jpg"

thresh = 100
maxval = 255
blockSize = 7
C = 3

img = cv.imread(path) 
img_gray = cv.cvtColor(img,cv.COLOR_BGR2GRAY)

plt.subplot(321)
plt.imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB))
plt.title(f'Original')
plt.axis('off')

plt.subplot(322)
plt.imshow(cv.cvtColor(img_gray, cv.COLOR_BGR2RGB))
plt.title(f'Gray Image')
plt.axis('off')

adaptiveMeanC = cv.adaptiveThreshold(img_gray, maxval, cv.ADAPTIVE_THRESH_MEAN_C, cv.THRESH_BINARY, blockSize=blockSize, C=C)
adaptiveMeanC = cv.cvtColor(adaptiveMeanC, cv.COLOR_BGR2RGB)
plt.subplot(323)
plt.imshow(adaptiveMeanC)
plt.title(f'adaptive Mean C thresh={thresh}, maxval={maxValue}')
plt.axis('off')

adaptiveGaussianC = cv.adaptiveThreshold(img_gray, maxval, cv.ADAPTIVE_THRESH_GAUSSIAN_C, cv.THRESH_BINARY, blockSize=blockSize, C=C)
adaptiveGaussianC = cv.cvtColor(adaptiveGaussianC,cv.COLOR_BGR2RGB)
plt.subplot(324)
plt.imshow(adaptiveGaussianC)
plt.title(f'adaptive Gaussian C thresh={thresh}, maxval={maxValue}')
plt.axis('off')


img_gray = cv.bilateralFilter(img_gray, 15, 25, 25)

adaptiveMeanC = cv.adaptiveThreshold(img_gray, maxval, cv.ADAPTIVE_THRESH_MEAN_C, cv.THRESH_BINARY, blockSize=blockSize, C=C)
adaptiveMeanC = cv.cvtColor(adaptiveMeanC, cv.COLOR_BGR2RGB)
plt.subplot(325)
plt.imshow(adaptiveMeanC)
plt.title(f'adaptive Mean C + bilateral filter')
plt.axis('off')

adaptiveGaussianC = cv.adaptiveThreshold(img_gray, maxval, cv.ADAPTIVE_THRESH_GAUSSIAN_C, cv.THRESH_BINARY, blockSize=blockSize, C=C)
adaptiveGaussianC = cv.cvtColor(adaptiveGaussianC,cv.COLOR_BGR2RGB)
plt.subplot(326)
plt.imshow(adaptiveGaussianC)
plt.title(f'adaptive Gaussian C + bilateral filter')
plt.axis('off')
```
